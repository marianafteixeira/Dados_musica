```python
#Importando a biblioteca Seaborn

import seaborn as sns
```

```python
#Importando a biblioteca Matplotlib

import matplotlib.pyplot as plt
```

```python
#Ordenando a média da popularidade dos gêneros musicais

df.groupby('Gênero')['Popularidade'].mean().sort_values(ascending = False)
```

```python
#Renomeando os gêneros musicais

Gênero = {
    'dfw rap': 'Rap',
    'electropop': 'Eletrônica',
    'trap music': 'Trap',
    'escape room': 'Escape Room',
    'pop house': 'Eletrônica',
    'reggaeton': 'Reggaeton',
    'latin': 'Música latina',
    'canadian hip hop': 'Hip Hop',
    'panamanian pop': 'Pop',
    'atl hip hop': 'Hip Hop',
    'country rap': 'Rap',
    'big room': 'Eletrônica',
    'brostep': 'Eletrônica',
    'r&b en espanol': 'R&B',
    'reggaeton flow': 'Reggaeton',
    'edm': 'Eletrônica',
    'pop': 'Pop',
    'dance pop': 'Pop',
    'australian pop': 'Pop',
    'boy band': 'Pop',
    'canadian pop': 'Pop'
    }

df['Gênero'] = df['Gênero'].replace(Gênero)
df['Gênero'].value_counts()
```

```python
#Ordenando a média da popularidade dos gêneros musicais

ordem = df.groupby('Gênero')['Popularidade'].mean().sort_values(ascending = False).index
ordem
```

```python
#Gráfico sobre a distribuição da popularidade das músicas

plt.figure(figsize=(15,8))
sns.histplot(data = df, x = 'Popularidade', bins = 50, kde = True)
plt.title('Distribuição da Popularidade')
plt.xlabel('Popularidade')
plt.ylabel('Frequência')
```

```python
#Boxplot sobre a duração das músicas

plt.figure(figsize = (10,6))
sns.boxplot(x = df['Duração'])
plt.title('Boxplot da duração das músicas')
plt.xlabel('Duração')
plt.show()
```

```python
#Importando a biblioteca Plotly

import plotly.express as px
```

```python
#Criando o gráfico interativo sobre a média da popularidade dos gêneros musicais

df_media_popularidade = df.groupby('Gênero')['Popularidade'].mean().reset_index()

fig = px.bar( df_media_popularidade,
    x='Gênero',
    y='Popularidade',
    title='Média da popularidade por gênero musical',
    labels={'Gênero': 'Gênero', 'Popularidade': 'Média da popularidade por gênero musical'},
    color='Gênero',
    category_orders={'Gênero': ordem}
)

fig.update_layout(xaxis_title='Gênero', yaxis_title='Média da popularidade por gênero musical')
fig.show()
```

```python
#Ordenando o gênero por sua dançabilidade

ordem2 = df.groupby('Gênero')['Dançabilidade'].mean().sort_values(ascending = False).index
ordem2
```

```python
#Criando o gráfico interativo sobre a média dos gêneros mais dançantes
#0 a 100, mostra o quão fácil é dançar a música

df_media_popularidade = df.groupby('Gênero')['Dançabilidade'].mean().reset_index()

fig = px.bar( df_media_popularidade,
    x='Gênero',
    y='Dançabilidade',
    title='Média dos gêneros mais dançantes',
    labels={'Gênero': 'Gênero', 'Dançabilidade': 'Média dos gêneros mais dançantes'},
    color='Gênero',
    category_orders={'Gênero': ordem2}
)

fig.update_layout(xaxis_title='Gênero', yaxis_title='Média dos gêneros mais dançantes')
fig.show()
```

```python
valencia_contagem = df['Gênero'].value_counts().reset_index()
valencia_contagem.columns = ['Gênero', 'Valencia']

fig = px.pie(valencia_contagem,
             names='Gênero',
             values='Valencia',
             title='Proporção da valencia por gênero',
             hole = 0.5

          )
fig.update_traces(textinfo='percent+label')

fig.show()
```
